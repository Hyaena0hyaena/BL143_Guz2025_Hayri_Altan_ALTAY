```python
# GÜN DEĞERLENDİRME #

# Burada "while" kullanarak bir döngü başlatıyoruz.
while True:

# Kullanıcan değer istediğimiz kısmı yazıyoruz.

  uyku_saat = input("Bu gün kaç saat uyudunuz? : ")
  ders_saat = input("Bugün kaç saat ders çalıştınız? : ")
  serbest_saat = input("Bu gün kaç saat serbest vakit geçirdiniz? : ")

# Kullanıcan istemiş olduğumuz değerlerin sayılsal olup olmadığını kontrol ediyoruz.

  if not uyku_saat.isdigit() or not ders_saat.isdigit() or not serbest_saat.isdigit():
    print("Sayısal bir veri giriniz lütfen.")
    continue

# Kullanıcan alıp kontrol ettiğimiz veriler "str" olduklarından dolayı bunları "int" değerlere çevirmek zorundayız.

  uyku_saat = int(uyku_saat)
  ders_saat = int(ders_saat)
  serbest_saat = int(serbest_saat)

# "int" Yapmış olduğumuz değerlerin negatif olup olmadığını kontrol ediyoruz.

  if uyku_saat < 0 or ders_saat < 0 or serbest_saat < 0:
    print("Girmiş olduğunuz değerler negatif olamaz.")
    continue

# Kullanıcının girmiş olduğu değerlerin toplamının 24 saat olması gerektiğinden dolayı bir "if" bloğu yazıyoruz.

  if uyku_saat + ders_saat + serbest_saat != 24:
    print("Girmiş olduğunuz değerler toplamı 24 saat olmalıdır.")
    continue

# Tüm koşulları sağladığımıza göre artık koda geçebiliriz.

# Bu kısımda kullanıcıdan aldığımız verileri değerlendiriyoruz.

  if uyku_saat < 6:
    uyku_deger = "Yetersiz uyku"
  elif uyku_saat <= 8:
    uyku_deger = "Normal uyku"
  else:
    uyku_deger = "Fazla uyku"


  if ders_saat < 6:
    ders_deger = "Normal çalışma günü"
  else:
    ders_deger = "Yoğun çalışma günü"

# Bu kısımda ders çalışma ve uyku saati verilerini değerlendiriyoruz.

  print("Uyku değeriniz: ", uyku_deger)
  print("Çalışma değeriniz: ", ders_deger)
  if uyku_deger == "Yetersiz uyku" and ders_deger == "Yoğun çalışma günü":
    print("Maalesef verimsiz bir gün geçiriyorsunuz.")
  else:
    print("Tebrikler dengeli bir gün geçiriyorsunuz.")
  break

# En son olarak "break" kullanarak döngüyü sonlandırıyoruz.
```
